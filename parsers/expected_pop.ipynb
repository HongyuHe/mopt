```python
import matplotlib.pyplot as plt
import numpy as np
import pprint
import os, sys
from collections import defaultdict
```

```python
output_fig_dir = "..\\figs\\"
if not os.path.isdir(output_fig_dir):
    os.mkdir(output_fig_dir)
```

```python
color_pallette = plt.get_cmap("Set1")
color_pallette
```

```python
# colors = [color_pallette(3), color_pallette(0), color_pallette(4), color_pallette(2)]
# colors = [color_pallette(1), color_pallette(2), color_pallette(3), color_pallette(4)]
colors = [color_pallette(1), color_pallette(3), color_pallette(2), color_pallette(4)]
line_styles = ["--", "--", "--", "--"]
marker_list = ["o", "X", "P", "s"]
```

```python
train = [47.5 * 100 / 190, 43.12 * 100  / 190, 38.3 * 100  / 190, 35.6 * 100  / 190, 32.5 * 100  / 190]
test = [23 * 100  / 190, 26.4 * 100  / 190, 26.1  * 100 / 190, 26.5 * 100  / 190, 27.2 * 100  / 190]
numSamples = [1, 2, 3, 4, 5]
```

```python
plt.rc('axes', linewidth=2)
plt.figure(figsize=(8, 3), dpi=300)
plt.plot(numSamples, train, label="Discovered", marker='o', color=colors[0],
             mec=colors[0], 
             markersize=10, 
             linewidth=4, 
             markeredgewidth=3,
             linestyle=line_styles[0])
plt.plot(numSamples, test, label="On 10 other random inst.", marker='X', color=colors[1],
             mec=colors[1], 
             markersize=10, 
             linewidth=4, 
             markeredgewidth=3,
             linestyle=line_styles[1])
plt.legend(fontsize=15.5, loc="upper right", ncol=1)
# plt.title("Pop on B4 (TimeLimit = 5m)", fontsize=16)
plt.xlabel("#random instances of Pop to compute expectation", fontsize=16)
plt.ylabel("norm advers gap (%)", fontsize=16)
# plt.xscale('log')
# plt.yticks(np.arange(0.25, 0.47, 0.03), fontsize=14)
plt.yticks(np.arange(12, 28, 5), fontsize=14)
plt.xticks(np.arange(1, 6, 1), fontsize=14)
plt.grid(which="major", zorder=-1, linestyle="--")
plt.savefig(output_fig_dir + "b4_expected_pop.jpg", bbox_inches="tight", format="jpg", dpi=300)
plt.savefig(output_fig_dir + "b4_expected_pop.pdf", bbox_inches="tight", format="pdf", dpi=300)
```
